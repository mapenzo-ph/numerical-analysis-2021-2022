### Reshaping and efficiency
Now we will see how to manipulate arrays, and reshape them, in order to maximixe the efficiency of our code.

```python
# declaring simple array from range object (yet another way)
a = array(range(100))
```

```python
# reshaping the array into a matrix of desired shape
A = a.reshape((10,10)) 
A
```

```python
# for the access to the elements, Python is row-major by default (order="C")
i = 2
j = 3
A[i,j] == a[10*i+j] # True for every i and j
```

```python
# change order in memory using option order="F" (column-major)
B = a.reshape((10,10), order="F")
B
```

```python
# proving order is col-major 
print(B[i,j] == a[10*i+j])
print(B[i,j] == a[10*j+i])
```

```python
# preferred order depends on the application
# remember to make switch ordering in loops
# to improve performance

# this is relevant for well performing numerical analysis
# each time you access memory, you want to do it locally

# C-style loop (columns are faster)
k = 0
correct = 0
wrong = 0
for i in range(10):
    for j in range(10):
        if A[i,j] == a[k]:
            correct += 1
        else:
            wrong += 1
        k += 1
        
correct, wrong
```

```python
# F-style loop (rows are faster)
k = 0
correct = 0
wrong = 0
for j in range(10):
    for i in range(10):
        if A[i,j] == a[k]:
            correct +=1
        else:
            wrong +=1
        k += 1
        
correct, wrong
```

## Polynomial interpolation
Now we move on to see how to perform polynomial interpolation.

### Building polynomial base functions
We start by constructing polynomial base functions (in three different ways), then we will use it to perform the interpolation.

```python
# set mesh to plot functions
N = 101
x = linspace(0,1,N)

# set number of base polynomials to plot
n = 5
```

```python
# plot monomial basis from x^0 to x^(n-1) (straightforward way)
for i in range(n):
    plot(x, x**i)
```

```python
# plot basis, but using matrix notation

# constructing matrix where the columns are vectors of x^i (evaluated on mesh)
# (in other words, row contains each polynomial evaluated on a point of mesh)

y = zeros((N,n))

for j in range(N):
    for i in range(n):
        y[j][i] = x[j]**i

plot(x,y)
```

```python
# plot basis, constructing y in one line
y = array([[xi**j for j in range(n)] for xi in x])

# Same thing, but even shorter
y = array([x**j for j in range(n)]).T

plot(x,y)
```

```python
plot(x,y)
```

Which is the best way?
The best one is the one that uses a single line, because it forces us to think of what we are doing. Furthermore, using only one line of code we are less likely to introduce bugs in the code.
(actually, both are wrong in some sense, since we are not using VECTORIZATION!)

### Writing a general polynomial
Now that we have seen how to plot the basis functions, we shall see how to construct a general polynomial using the basis functions. As an example we try to plot the following polynomial:

$$
g(x) = -2*x^2 + 2*x +1
$$

To do this we start by building the vector representation of the function, i.e the vector of the coefficients of our polynomial with respect to the basis functions, then we can use the matrix y we have built befor to plot the function.

```python
# construct vector representation -> p[i] = coeff of x^i
p = zeros(n) 
p[0] = 1
p[1] = +2 # coefficient x
p[2] = -2 # coefficient x^2
# the rest are zeroes
```

```python
# morally we want the dot product g = y*p
# to perform dot prod we need to write
g = y.dot(p)
```

```python
# plotting our function
plot(x,g)
```

### Interpolation of the Runge function
Now that we know how to write a general polynomial, we can proceed to perform interpolation of a general function. We choose the Runge function, which we have already defined at the beginning of the notebook.

The first thing we should do is setting up our interpolation points. Here we take equispaced points. After that, we can create the matrix $V_{ij}$ whose elements are defined as

$$ V_{ij} = v_j(q_i) $$

where $\{v_j\}_{j=0}^{n-1}$ are the polynomial basis functions. The interpolation of our function is then given by

$$
I^n[f] (x) = p^i v_i(x)
$$

where $p_i$ is the vector of coefficients of the interpolated polynomial,computed as

$$ p^i = V^{ij} f(q_j)$$

(NOTE: $V^{ij}$ here denotes the element of the inverse matrix of $V_{ij}$)

```python
# defining target function
def runge(x):
    return 1/(1+100*(x-.5)**2)
```

```python
# setting interpolation points
Nq = 5
q = linspace(0,1,Nq)
```

```python
# setting up the V matrix
V = array([q**j for j in range(n)]).T 
V.shape
```

```python
# computing inverse of V
Vinv = linalg.inv(V)
```

```python
# computing vector f(q_i)
r = runge(q)
```

```python
# computing the p vector (interpolation polyomial coefficients)
p = Vinv.dot(r)
```

```python
# plotting interploated function (as we did for g(x) before)
plot(x, runge(x), label='runge')
plot(x, y.dot(p), label='interp.')
plot(q, runge(q), 'ro')
legend()
```

```python
# building function that performs interpolation on given function
def interpolate(q,x,function):
    """
    Use the q points to construct interpolation of function on
    points x.
    
    Return the polynomial interpolation function evaluated at x.
    """
    
    # interpolation (independent of x)
    n = len(q)
    V = array([q**j for j in range(n)]).T  
    r = function(q)
    p = linalg.solve(V,r) # equals to Vinv.dot(r)
    
    # matrix for plot
    Y = array([x**j for j in range(n)]).T
    
    return Y.dot(p)
```

```python
# interpolation result when varying the number of interpolation points
for n in range(5,10):
    q = linspace(0,1,n)
    plot(x,runge(x))
    plot(x, interpolate(q,x,runge))
    plot(q,runge(q), 'ro')
    show()
```

### Exercise 1
How does the interpolation error $E(n)$ grow when we increase the number of interpolation points?
Use $
E(n) := \max \, |\text{runge}(x) - \text{interpolation}(q,x,\text{runge})|
$ 

```python
# function that computes interpolation error given set of points and target function
def interpolation_error(q):
    return max(abs(runge(x) - interpolate(q,x,runge)))
```

```python
# define array with different numbers of interpolation points
ns = array(range(5,20))
```

```python
# compute vector of errors
errs = array([interpolation_error(linspace(0,1,np)) for np in ns])
```

```python
# plotting interpolation errors
plt.xlabel('Number of interpolation points')
plt.ylabel('Interpolation error')
plot(ns, errs)
```

### Exercise 2
Now fix the number of interpolation point $n=10$ and find algorithmically the distribution of points that minimizes the error.

```python
# reinitialize mesh and q
x = linspace(0,1, 1001)
q = linspace(0,1,10)
```

```python
# Chebyshev nodes on [0,1]
cheb = array([0.5+0.5*cos((2*k-1)*pi/(2*len(q))) for k in range(len(q)+1)])
cheb
```

```python
# function that computes gradient of error wrt q points
def grad_q_error(q, delta):
    D = delta*eye(len(q))
    qp = array([q+D[i,:] for i in range(len(q))])
    qm = array([q-D[i,:] for i in range(len(q))])
    errs = array([(interpolation_error(qp[i,:]) - interpolation_error(qm[i,:]))/(2*delta) for i in range(len(q))])
    
    return errs/linalg.norm(errs)
```

```python
# gradient descent to find best q

# initial guess for q
q = linspace(0,1,10)

# initial value for interpolation error
e = interpolation_error(q)

# gradient descent parameter
alpha = 0.07

diff = 1

# iteratively modifying q (gradient descent)
while diff > 0.00001:
    e_old = e
    p = grad_q_error(q, 0.00005)    
    q = q - alpha*p    
    e = interpolation_error(q)
    diff = e - e_old
```

```python
plot(x,runge(x))
plot(x, interpolate(q,x,runge))
plot(q, runge(q), 'ro')
plot(cheb, runge(cheb), 'o')

show()
```
